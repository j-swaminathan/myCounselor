```python

import plotly.graph_objects as go
import plotly.offline as pyo


categories = ['Health & Medicine', 'Agriculture & Sciences', 'Arts & Communication', 'Enginnering & Technology', 'Buisness & Management','Public Service']
categories = [*categories, categories[0]]


'''
Enter values of the survey to generate Spider graph.
'''

#participant_1 = [6, 5, 5, 5, 3, 10]
#participant_1_ridhi = [11, 10.4, 10.2, 9.4, 9.6, 10.2]
#participant_1_kaashu = [13, 14, 15.2, 11, 13, 15]
#participant_1 = [203,182,175,155,168,167]
participant_1=[9.8,7,8,6.6,5.6,12]


participant_1 = [*participant_1, participant_1[0]]



fig = go.Figure(
    data=[
        go.Scatterpolar(r=participant_1, theta=categories, name='Alexandera')
         ],
    layout=go.Layout(
        title=go.layout.Title(text='Participant'),
        polar={'radialaxis': {'visible': True}},
        showlegend=True
    )
)

pyo.plot(fig)
```
